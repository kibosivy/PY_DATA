## VISUALIZATION
-barplots: vertical,horizontal,stacked and grouped
-customization
-scatter plots and color bars

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
drinks = pd.read_csv(r"C:\Users\USER\desktop\PY_DATA\drinks.csv")
drinks.head()
```

```python
# bar plot-visualize the counts of categorical variables
# beer servings per country
plt.bar(drinks['country'], drinks['beer_servings'], color='pink', edgecolor='black')
plt.show()
```

```python
top_10 = drinks.sort_values(by='beer_servings', ascending=False).head(10)
top_10
```

```python
plt.bar(top_10['country'], top_10['beer_servings'], color='pink', edgecolor='black')


plt.title('Beer Servings per person in Top 10')
plt.xticks(rotation=90)
plt.yticks(rotation=30)
plt.xlabel('Country')
plt.ylabel('Average beer servings per person')
plt.show()
```

```python
# advanced customization- adding anotations
plt.bar(top_10['country'], top_10['beer_servings'], color='pink', edgecolor='black')


plt.title('Beer Servings per person in Top 10')
plt.xticks(rotation=90)
plt.yticks(rotation=30)
plt.xlabel('Country')
plt.ylabel('Average beer servings per person')
plt.gca().axes.get_yaxis().set_visible(False)

# plt.text()- you have to define for it in an x and y coordinate
for x, y in enumerate (top_10['beer_servings']):
    plt.text(x, y+5, str(y), ha='center')

plt.show()
```

### Grouped Bar Chart

```python
plt.figure(figsize=(10,6))

bar_width = 0.5
x = np.arange(len(top_10['country']))

plt.bar(x-bar_width/2, top_10['beer_servings'], bar_width, label='Beer', color='pink')
plt.bar(x+bar_width/2, top_10['spirit_servings'], bar_width, label='Beer', color='green')

plt.xticks(x, top_10['country'], rotation=30)
plt.title('Beer & Spirit Servings per person in top 10 countries')

for x, y in enumerate (top_10['beer_servings']):
    plt.text(x, y+5, str(y), ha='right')

for x, y in enumerate (top_10['spirit_servings']):
    plt.text(x, y+5, str(y), ha='left')

plt.show()
```

## Assignment: 
plot a visualization to compare the beer, spirits and wine servings of Uganda, Tanzania, Kenya

```python
three_countries = ['Kenya', 'Uganda', 'Tanzania']
filtered_df = drinks[drinks['country'].isin(three_countries)]

bar_width = 0.25
x = np.arange(len(filtered_df))

plt.figure(figsize=(10, 6))

plt.bar(x - bar_width, filtered_df['beer_servings'], width=bar_width, label='Beer', color='blue')
plt.bar(x, filtered_df['spirit_servings'], width=bar_width, label='Spirits', color='grey')
plt.bar(x + bar_width, filtered_df['wine_servings'], width=bar_width, label='Wine', color='orange')

plt.xlabel('Country')
plt.ylabel('Servings')
plt.title('Alcohol Consumption by Type in Kenya, Uganda, and Tanzania')
plt.xticks(x, filtered_df['country'])  # label x-axis with country names
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# stacked bar chart
plt.figure(figsize=(10,6))

plt.bar(top_10['country'], top_10['beer_servings'], color='purple')
plt.bar(top_10['country'], top_10['spirit_servings'], bottom=top_10['beer_servings'], color='yellow')
plt.bar(top_10['country'], top_10['wine_servings'], bottom=top_10['beer_servings'] + top_10['spirit_servings'], color='red')

plt.show
```

## Plotting multiple plots on the same figure

```python
# subplots
fig,axs=plt.subplots(1,3, figsize=(12,6))

axs[0].bar(top_10['country'], top_10['beer_servings'])
axs[1].bar(top_10['country'], top_10['wine_servings'])
axs[2].bar(top_10['country'], top_10['spirit_servings'])
```

```python
# subplots
fig,axs=plt.subplots(2,2, figsize=(12,6))

axs[0,0].bar(top_10['country'], top_10['beer_servings'])
axs[0,1].bar(top_10['country'], top_10['wine_servings'])
axs[1,0].bar(top_10['country'], top_10['spirit_servings'])
axs[1,1].bar(top_10['country'], top_10['total_litres_of_pure_alcohol'])
```
