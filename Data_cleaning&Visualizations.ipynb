## The Data Analysis Pipeline

1. Clean data
    - Handling missingness
    - Handling outliers
    - Duplicates
    - Addressing data type mismatches
2. Univariate Analysis- line plots, bar charts, kde plots, histograms, box plots, violin plots
3. Bivaraiate Analysis- identify relationships- grouped bar charts, scatter plots,
4. Multivariate Analysis- correlation matrix, heatmap

## Identifying and Handling Outliers

```python
numeric_cols=train.select_dtypes(include=np.number).columns.tolist()
numeric_cols
```

```python
cols=['meals_served','kitchen_staff','temperature_C','humidity_percent','past_waste_kg','food_waste_kg']
num_df=train[cols]

fig = plt.figure(figsize=(15,30))
for i in range(len(cols)):
    ax= fig.add_subplot(9, 5, i+1)
    sns.boxplot(x=num_df.iloc[:, i],ax=ax)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,5))

sns.boxplot(x='temperature_C',data=num_df)

plt.show()
```

```python
temp_df=train['temperature_C']
temp_df.head()
```

```python
Q3=np.quantile(train['temperature_C'],0.75)
Q1=np.quantile(train['temperature_C'],0.25)

IQR=Q3-Q1

lower_range=Q1-(1.5*IQR)
upper_rannge=Q3+ (1.5*IQR)
```

```python
clean_df=train[(train['temperature_C']>= lower_range) & (train['temperature_C']<=upper_rannge)]
clean_df.shape
```
